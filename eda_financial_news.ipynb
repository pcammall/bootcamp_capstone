```python
temp_df = df['text']
terms = {}
for x in temp_df:
    for y in x.split():
        if terms.get(y):
            terms[y] = terms[y] + 1
        else:
            terms.setdefault(y, 1)
        
        

for key, value in sorted(terms.items(), key=itemgetter(1), reverse=True):
    print(key, value)
```
