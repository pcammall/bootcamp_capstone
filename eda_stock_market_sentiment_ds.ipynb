```python
#convert each text string to list to get count of which terms appear the most
temp_df = df['Text']
terms = {}
for x in temp_df:
    for y in x.split():
        if terms.get(y):
            terms[y] = terms[y] + 1
        else:
            terms.setdefault(y, 1)
        
        
from operator import itemgetter
for key, value in sorted(terms.items(), key=itemgetter(1), reverse=True):
    print(key, value)

    
#Not terribly hepful at first. The most common terms used are article words (the/in/a/it type of stuff.)
#standout is AAP at 884. Either means Automated Accredited Clearing House Professional, or the company Advanced Auto Parts
#not sure how this would affect sentiment analysis. Might ony be important if i focus on Advanced Auto Parts
#Some common tickers are NFX (nuflorix?), RT (right tunneling?), and Amazon. The first two are penny stocks if correct
```
